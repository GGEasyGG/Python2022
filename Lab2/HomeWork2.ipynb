**Задание 2 (1 балл):** Опишите класс комплексных чисел. У пользователя должна быть возможность создать его объект на основе числа и в алгебраической форме, и в полярной. Класс должен поддерживать основные математические операции (+, -, *, /) за счет перегрузки соответствующих магических методов. Также он должен поддерживать возможность получить число в алгебраической и полярной форме. Допускается использование модуля `math`.

```python
import math

class ComplexNumber:
  def __init__(self, type_of_complex, a = 0.0, b = 0.0):
    self.__re = a
    self.__im = b
    self.__type_of_complex = type_of_complex   

  def __repr__(self):
    return 'ComplexNumber({}, {}, {})'.format(self.__re, self.__im, self.__type_of_complex)

  def __str__(self):
    if(self.__type_of_complex == "alg"):
      if(self.__im < 0):
        sim = "-"
      else:
        sim = "+"

      re = self.__re
      im = self.__im

      if((0.0 < re < 1.0e-5) or (-1.0e-5 < re < 0.0)):
        re = 0
      if((0.0 < im < 1.0e-5) or (-1.0e-5 < im < 0.0)):
        im = 0

      if re != 0 and im != 0:
        if(abs(im) == 1):
          return "%.5f" % re + sim + str(abs(im)).split("1")[0] + "i"
        else:
          return "%.5f" % re + sim + "%.5f" % abs(im) + "i"
      elif im != 0:
        if(abs(im) == 1):
          return str(im).split("1")[0] + "i"
        else:
          return "%.5f" % im + "i"
      else:
        return "%.5f" % re
    else:
      return "r = " + "%.5f" % self.__re + ", phi = " + "%.5f" % self.__im

  def __add__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      re_sum = first.__re + other
      im_sum = first.__im
    elif(isinstance(other, ComplexNumber)):
      second = other.to_alg()
      re_sum = first.__re + second.__re
      im_sum = first.__im + second.__im
    else:
      raise TypeError('Unsupported type')

    complex_alg = ComplexNumber("alg", re_sum, im_sum)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()

  def __radd__(self, other):
    return self + other

  def __sub__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      re_sub = first.__re - other
      im_sub = first.__im
    elif(isinstance(other, ComplexNumber)):
      second = other.to_alg()
      re_sub = first.__re - second.__re
      im_sub = first.__im - second.__im
    else:
      raise TypeError('Unsupported type')

    complex_alg = ComplexNumber("alg", re_sub, im_sub)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()

  def __rsub__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      re_sub = other - first.__re
      im_sub = first.__im
    elif(isinstance(other, ComplexNumber)):
      second = other.to_alg()
      re_sub = second.__re - first.__re 
      im_sub = second.__im - first.__im 
    else:
      raise TypeError('Unsupported type')

    complex_alg = ComplexNumber("alg", re_sub, im_sub)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()
        
  def __mul__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      re_mul = other * first.__re
      im_mul = other * first.__im
    elif(isinstance(other, ComplexNumber)):
      second = other.to_alg()
      re_mul = first.__re * second.__re - first.__im * second.__im
      im_mul = first.__im * second.__re + first.__re * second.__im 
    else:
      raise TypeError('Unsupported type')
  
    complex_alg = ComplexNumber("alg", re_mul, im_mul)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()

  def __rmul__(self, other):
    return self * other

  def __truediv__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      if(other == 0):
        raise ZeroDivisionError('Division by 0')
      re_div = first.__re / other
      im_div = first.__im / other
    elif(isinstance(other, ComplexNumber)):
      if((other.__im == 0) and (other.__re == 0)):
        raise ZeroDivisionError('Division by 0')
      second = other.to_alg()
      a = first.__re
      b = first.__im
      c = second.__re
      d = second.__im
      zz = c**2 + d**2
      re_div = (a * c + b * d) / zz
      im_div = (b * c - a * d) / zz 
    else:
      raise TypeError('Unsupported type')
    
    complex_alg = ComplexNumber("alg", re_div, im_div)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()

  def __rtruediv__(self, other):
    first = self.to_alg()

    if(isinstance(other, int) or isinstance(other, float)):
      a = other
      b = 0
      c = first.__re
      d = first.__im
      zz = c**2 + d**2
      re_div = (a * c + b * d) / zz
      im_div = (b * c - a * d) / zz 
    elif(isinstance(other, ComplexNumber)):
      second = other.to_alg()
      a = second.__re
      b = second.__im
      c = first.__re
      d = first.__im
      zz = c**2 + d**2
      re_div = (a * c + b * d) / zz
      im_div = (b * c - a * d) / zz 
    else:
      raise TypeError('Unsupported type')
    
    complex_alg = ComplexNumber("alg", re_div, im_div)

    if(self.__type_of_complex == "alg"):
      return complex_alg
    else:
      return complex_alg.to_pol()

  def to_alg(self):
    if(self.__type_of_complex == "pol"):
      re = self.__re * math.cos(math.radians(self.__im))
      im = self.__re * math.sin(math.radians(self.__im))
      return ComplexNumber("alg", re, im)
    else:
      return self

  def to_pol(self):
    if(self.__type_of_complex == "alg"):
      if self.__re == 0 and self.__im == 0:
        raise ValueError("Untranslatable number")

      r = (self.__re**2 + self.__im**2)**0.5
      if((self.__re > 0.0) and (self.__im > 0.0)):
        phi = math.degrees(math.atan(self.__im / self.__re))
      elif((self.__re < 0.0) and (self.__im > 0.0)):
        phi = 180.0 + math.degrees(math.atan(self.__im / self.__re))
      elif((self.__re < 0.0) and (self.__im < 0.0)):
        phi = 180.0 + math.degrees(math.atan(self.__im / self.__re))
      elif((self.__re > 0.0) and (self.__im < 0.0)):
        phi = 360.0 + math.degrees(math.atan(self.__im / self.__re))
      elif(self.__re == 0.0):
        if(self.__im > 0.0):
          phi = math.degrees(math.atan(math.inf))
        else:
          phi = 180.0 + math.degrees(math.atan(math.inf))
      elif(self.__im == 0.0):
        if(self.__re > 0.0):
          phi = math.degrees(math.atan(0.0))
        else:
          phi = 180.0 + math.degrees(math.atan(0.0))

      return ComplexNumber("pol", r, phi)
    else:
      return self

  def get_pol(self):
    print(self.to_pol())

  def get_alg(self):
    print(self.to_alg())
```

**Задание 3 (2 балла):** Опишите класс для векторов в N-мерном пространстве. В качестве основы используйте список значений координат вектора, задаваемый `list`. Обеспечьте поддержку следующих операций: сложение, вычитание (с созданием нового вектора-результата), скалярное произведение, косинус угла, евклидова норма. Все операции, которые можно перегрузить с помощью магических методов, должны быть реализованы именно через них. Класс должен производить проверку консистентности аргументов для каждой операции и в случаях ошибок выбрасывать исключение `ValueError` с исчерпывающим объяснением ошибки.

```python
class Vector:
  def __init__(self, vector_values_list):
    if(isinstance(vector_values_list, list)):
      if(len(vector_values_list) != 0):
        if(all((isinstance(x, int) or isinstance(x, float))for x in vector_values_list)):
          self.__vect = vector_values_list.copy()
          self.__lenght = len(vector_values_list)
        else:
          raise ValueError("Vector elements can only be int or float")
      else:
        raise ValueError("You cannot create a vector from an empty list")
    else:
      raise TypeError("A vector can be created only from a list")

  def __repr__(self):
    return 'Vector({})'.format(self.__vect)

  def __str__(self):
    return "(" + ", ".join(map(str, self.__vect)) + ")"

  def __add__(self, other):
    if(not isinstance(other, Vector)):
      raise TypeError("The argument is not a vector")

    if(self.__lenght == other.__lenght):
      result_vect = []
      for i, j in zip(self.__vect, other.__vect):
        result_vect.append(i + j)
    
      return Vector(result_vect)
    else:
      raise ValueError("Vectors have different dimensions")

  def __radd__(self, other):
    return self + other

  def __sub__(self, other):
    if(not isinstance(other, Vector)):
      raise TypeError("The argument is not a vector")

    if(self.__lenght == other.__lenght):
      result_vect = []
      for i, j in zip(self.__vect, other.__vect):
        result_vect.append(i - j)
    
      return Vector(result_vect)
    else:
      raise ValueError("Vectors have different dimensions")

  def __rsub__(self, other):
    return self - other
      
  def scalar_product(self, other):
    if(not isinstance(other, Vector)):
      raise TypeError("The argument is not a vector")

    if(self.__lenght == other.__lenght):
      sum = 0
      for i, j in zip(self.__vect, other.__vect):
        sum += i * j
    
      return sum
    else:
      raise ValueError("Vectors have different dimensions")

  def e_norm(self):
    sum = 0
    for i in self.__vect:
      sum += i**2
  
    return sum**0.5

  def angle_cos(self, other):
    if(not isinstance(other, Vector)):
      raise TypeError("The argument is not a vector")

    if(self.__lenght == other.__lenght):
      return self.scalar_product(other) / (self.e_norm() * other.e_norm())
    else:
      raise ValueError("Vectors have different dimensions")
```

```python
def calls_counter(func):
  def __counter(*args, **kwargs):
    __counter.count += 1
    print("This function was called %d times" % __counter.count)
    return func(*args, **kwargs)
    
  __counter.count = 0
  return __counter
```

```python
def copy_class_attrs(cls_2):
  def _decorate(cls_1):
    for f in cls_2.__dict__:
      if not f.startswith("_"):
        if(f in cls_1.__dict__):
          setattr(cls_1, f + "_new", getattr(cls_2, f))
        else:
          setattr(cls_1, f, getattr(cls_2, f))
    return cls_1
  return _decorate
```

В реализации математических операций запрещается пользоваться любыми функциями, требующими использования оператора `import`.

```python
class CopyMatrix:
  def copy_matrix(self):
    return Matrix(self.matrix.copy())

class LoadMatrix:
  def load_matrix_from_bin(self, filename):
    file = open(filename, "rb")

    lines = [line for line in file.readlines()]

    file.close()

    matrix = []

    for item in lines:
      rows = []
      line = item.decode()[:-2]
      
      line = line.split(" ")

      for elem in line:
        try:
          rows.append(float(elem))
        except ValueError:
          rows.append(elem)

      matrix.append(rows)

    return Matrix(matrix)

  def load_matrix_from_txt(self, filename):
    file = open(filename, "r")

    lines = [line.strip() for line in file.readlines()]

    file.close()

    matrix = []

    for item in lines:
      rows = []
      line = item.split(" ")

      for elem in line:
        try:
          rows.append(float(elem))
        except ValueError:
          rows.append(elem)

      matrix.append(rows)

    return Matrix(matrix)

class SaveMatrix:
  def save_matrix_as_binary(self, filename):
    with open(filename, 'wb') as testfile:
      for row in self.matrix:
        for item in row:
          s = str(item) + ' '
          testfile.write(s.encode())
        testfile.write('\n'.encode())

  def save_matrix_as_txt(self, filename):
    with open(filename, 'w') as testfile:
      for row in self.matrix:
        testfile.write(' '.join([str(a) for a in row]) + '\n')

class Matrix(CopyMatrix, LoadMatrix, SaveMatrix):
  def __init__(self, *args):
    if(len(args) == 1):
      if(isinstance(args[0], list)):
        if(all(isinstance(elem, (int, float)) for elem in args[0])):
          self.matrix = [args[0].copy()]
        elif(all(isinstance(elem, list) for elem in args[0])):
          if(len(set(len(elem) for elem in args[0])) == 1):
            if(all(isinstance(elem, (int, float)) for elem in sum(args[0], []))):
              self.matrix = args[0].copy()
            else:
              raise ValueError("Matrix elements must be of type int or float")
          else:
            raise ValueError("All rows of the matrix must be of the same length")
        else:
          raise ValueError("Matrix elements must be of type int or float")

        self.number_of_rows = len(self.matrix)
        self.number_of_columns = len(self.matrix[0])
      else:
        raise TypeError("The matrix can be created only based on a list, or by specifying the number of rows and columns with or without the default element")
    
    elif(len(args) == 2):
      if(all(isinstance(elem, int) for elem in args)):
        self.matrix = [[0 for _ in range(args[1])] for _ in range(args[0])]
        self.number_of_rows = args[0]
        self.number_of_columns = args[1]
      else:
        raise TypeError("The number of rows and columns must be of type int")

    elif(len(args) == 3):
      if(all(isinstance(elem, int) for elem in args[:2])):
        if(isinstance(args[2], (int, float))):
          self.matrix = [[args[2] for _ in range(args[1])] for _ in range(args[0])]
          self.number_of_rows = args[0]
          self.number_of_columns = args[1]
        else:
          raise TypeError("The default matrix element must be of type int or float")
      else:
        raise TypeError("The number of rows and columns must be of type int")

    else:
      raise Exception("It is not possible to create a matrix from a given number of parameters")

  def __repr__(self):
    return 'Matrix({})'.format(self.matrix)

  def __str__(self):
    return '\n'.join('\t'.join(map(str, row)) for row in self.matrix)

  def __getitem__(self, pos):
    i, j = pos
    return self.matrix[i][j]

  def __setitem__(self, pos, value):
    i, j = pos

    if(isinstance(value, (int, float))):
      self.matrix[i][j] = value
    else:
      raise TypeError("Matrix elements must be of type int or float")

  def __add__(self, other):
    if(not isinstance(other, Matrix)):
      raise TypeError("Only the matrix can be added to the matrix")
    else:
      if((self.number_of_rows == other.number_of_rows) and (self.number_of_columns == other.number_of_columns)):
        matrix = Matrix(self.number_of_rows, self.number_of_columns)
        for i in range(self.number_of_rows):
          for j in range(self.number_of_columns):
            matrix[i, j] = self.matrix[i][j] + other.matrix[i][j]
        return matrix
      else:
        raise ValueError("You can only add matrices of the same size")

  def __radd__(self, other):
    return self + other

  def __sub__(self, other):
    if(not isinstance(other, Matrix)):
      raise TypeError("Only matrices can be subtracted")
    else:
      if((self.number_of_rows == other.number_of_rows) and (self.number_of_columns == other.number_of_columns)):
        matrix = Matrix(self.number_of_rows, self.number_of_columns)
        for i in range(self.number_of_rows):
          for j in range(self.number_of_columns):
            matrix[i, j] = self.matrix[i][j] - other.matrix[i][j]
        return matrix
      else:
        raise ValueError("You can only subtract matrices of the same size")

  def __rsub__(self, other):
    return self - other

  def __mul__(self, other):
    if(isinstance(other, Matrix)):
      raise TypeError("The matrices are multiplied using the dot method")
    else:
      if(isinstance(other, (int, float))):
        matrix = Matrix(self.number_of_rows, self.number_of_columns)
        for i in range(self.number_of_rows):
          for j in range(self.number_of_columns):
            matrix[i, j] = self.matrix[i][j] * other
        return matrix
      else:
        raise TypeError("Matrices can be multiplied only with the int or float type")

  def __rmul__(self, other):
    return self * other

  def __truediv__(self, other):
    if(isinstance(other, Matrix)):
      raise TypeError("You cannot divide matrices")
    else:
      if(isinstance(other, (int, float))):
        matrix = Matrix(self.number_of_rows, self.number_of_columns)
        for i in range(self.number_of_rows):
          for j in range(self.number_of_columns):
            matrix[i, j] = self.matrix[i][j] / other
        return matrix
      else:
        raise TypeError("The division operation with the matrix is performed only with the int or float type")

  def __rtruediv__(self, other):
    raise TypeError("You can't divide a number into a matrix")

  def dot(self, other):
    if(isinstance(other, Matrix)):
      if(self.number_of_columns == other.number_of_rows):
        matrix = Matrix(self.number_of_rows, other.number_of_columns)
        for i in range(self.number_of_rows):
          for j in range(other.number_of_columns):
            for k in range(self.number_of_columns):
              matrix[i, j] =  matrix[i, j] + self.matrix[i][k] * other.matrix[k][j]

        return matrix
      else:
        raise TypeError("It is impossible to multiply matrices of specified sizes")
    else:
      raise TypeError("In this method, only matrices can be multiplied")

  def T(self):
    matrix = Matrix(self.number_of_columns, self.number_of_rows)

    for i in range(self.number_of_rows):
      for j in range(self.number_of_columns):
        matrix[j, i] = self.matrix[i][j]

    return matrix

  def tr(self):
    if(self.number_of_rows == self.number_of_columns):
      summ = 0

      for i in range(self.number_of_rows):
        summ += self.matrix[i][i]

      return summ
    else:
      print("Error: Only the square matrix has a trace")
      return None

  def det(self):
    if(self.number_of_rows == self.number_of_columns):
      det = 1

      triang_matrix = self.matrix.copy()

      for i in range(self.number_of_rows - 1):
        flag = True
        if(triang_matrix[i][i] == 0.0):
          flag = False
          for r in range(i + 1, self.number_of_rows):
            if(triang_matrix[r][i] != 0.0):
              flag = True
              tmp = triang_matrix[i]
              triang_matrix[i] = triang_matrix[r]
              triang_matrix[r] = tmp
              det *= -1
              break

        if(flag == False):
          continue
        for j in range(i + 1, self.number_of_rows):
          triang_matrix[j] = list(map(lambda x: 0.0 if (-1e-8 < x < 1e-8) else x,
            [triang_matrix[j][k] - (triang_matrix[j][i] * triang_matrix[i][k]) / triang_matrix[i][i] for k in range(self.number_of_rows)]))

      for i in range(self.number_of_rows):
        det *= triang_matrix[i][i]

      return det
    else:
      print("Error: Only a square matrix has a determinant")
      return None
```

Добавки могут включаться в чашку в любом количестве и порядке. Добавление новых типов кофе и добавок не должно требовать изменения существующего кода.

```python
class CoffeeInterface():
  def calculate_cost(self):
    raise NotImplementedError

  def coffee_name(self):
    raise NotImplementedError

class Latte(CoffeeInterface):
  def __init__(self, *args):
    if(len(args) != 0):
      raise Exception("Coffee can occur in a cup of coffee only once")

  def calculate_cost(self):
    return 10.0

  def coffee_name(self):
    return "Latte"

class Cappuccino(CoffeeInterface):
  def __init__(self, *args):
    if(len(args) != 0):
      raise Exception("Coffee can occur in a cup of coffee only once")

  def calculate_cost(self):
    return 15.0

  def coffee_name(self):
    return "Cappuccino"

class Americano(CoffeeInterface):
  def __init__(self, *args):
    if(len(args) != 0):
      raise Exception("Coffee can occur in a cup of coffee only once")

  def calculate_cost(self):
    return 20.0

  def coffee_name(self):
    return "Americano"

class Sugar(CoffeeInterface):
  def __init__(self, coffee = None):
    if(coffee is None):
      raise Exception("The first element of a cup of coffee should be the coffee itself")

    self.coffee = coffee
  
  def calculate_cost(self):
    return self.coffee.calculate_cost() + 1.0

  def coffee_name(self):
    return self.coffee.coffee_name() + " + sugar"

class Cream(CoffeeInterface):
  def __init__(self, coffee = None):
    if(coffee is None):
      raise Exception("The first element of a cup of coffee should be the coffee itself")

    self.coffee = coffee
  
  def calculate_cost(self):
    return self.coffee.calculate_cost() + 2.0

  def coffee_name(self):
    return self.coffee.coffee_name() + " + cream"

class Cardamom(CoffeeInterface):
  def __init__(self, coffee = None):
    if(coffee is None):
      raise Exception("The first element of a cup of coffee should be the coffee itself")

    self.coffee = coffee
  
  def calculate_cost(self):
    return self.coffee.calculate_cost() + 3.0

  def coffee_name(self):
    return self.coffee.coffee_name() + " + cardamom"

class Foam(CoffeeInterface):
  def __init__(self, coffee = None):
    if(coffee is None):
      raise Exception("The first element of a cup of coffee should be the coffee itself")

    self.coffee = coffee
  
  def calculate_cost(self):
    return self.coffee.calculate_cost() + 4.0

  def coffee_name(self):
    return self.coffee.coffee_name() + " + foam"

class Syrup(CoffeeInterface):
  def __init__(self, coffee = None):
    if(coffee is None):
      raise Exception("The first element of a cup of coffee should be the coffee itself")

    self.coffee = coffee
  
  def calculate_cost(self):
    return self.coffee.calculate_cost() + 5.0

  def coffee_name(self):
    return self.coffee.coffee_name() + " + syrup"
```

```python
try:
  coffee2 = Cream(Americano(Sugar(Latte())))
except Exception as e:
  print(e)

try:
  coffee2 = Americano(Latte(Cappuccino()))
except Exception as e:
  print(e)

try:
  coffee2 = Cream(Sugar(Latte(Syrup())))
except Exception as e:
  print(e)

try:
  coffee2 = Cream(Foam(Sugar()))
except Exception as e:
  print(e)

coffee2 = Syrup(Sugar(Cardamom(Foam(Cream(Cappuccino())))))
print(coffee2.coffee_name() + ": " + str(coffee2.calculate_cost()))

coffee3 = Cream(Sugar(Latte()))
print(coffee3.coffee_name() + ": " + str(coffee3.calculate_cost()))

coffee4 = Syrup(Sugar(Cardamom(Sugar(Cream(Cream(Cream(Cardamom(Sugar(Cream(Syrup(Foam(Syrup(Americano())))))))))))))
print(coffee4.coffee_name() + ": " + str(coffee4.calculate_cost()))

coffee1 = Latte()
print(coffee1.coffee_name() + ": " + str(coffee1.calculate_cost()))
```
